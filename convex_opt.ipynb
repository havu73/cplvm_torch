The general problem is:
Find the maximum of f(x) = (x-3)^2 
contraints: x >= 5

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the function f(x) and the line y = x
f_x = lambda x: (x - 3)**2
y_x = lambda x: x

# Generate x values
x_values = np.linspace(0, 6, 400)

# Plot f(x) and y = x
plt.figure(figsize=(10, 6))
plt.plot(x_values, f_x(x_values), label=r'$f(x) = (x-3)^2$')
plt.plot(x_values, y_x(x_values), label=r'$y = x$', linestyle='--')

# Title and legend
plt.title('Plot of $f(x) = (x-3)^2$ and $y = x$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.legend()

plt.grid(True)
plt.show()
```
